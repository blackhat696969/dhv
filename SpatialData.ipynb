location of my home


```python
import folium

m = folium.Map(location=[29.1869408823662, 79.48816129130711]
               )

folium.Marker(
    location=[29.1869408823662, 79.48816129130711],
    popup='My Home',
    icon=folium.Icon()
).add_to(m)

m
```

world map


```python


import folium
from folium.plugins import MarkerCluster
from IPython.display import display, Image


world_map = folium.Map(location=[0, 0], zoom_start=2)


places = [
    {"name": "New York City", "coords": [40.7128, -74.0060], "image_url": "https://source.unsplash.com/400x300/?newyork"},
    {"name": "London", "coords": [51.5074, -0.1278], "image_url": "https://source.unsplash.com/400x300/?london"},
    {"name": "Tokyo", "coords": [35.6895, 139.6917], "image_url": "https://source.unsplash.com/400x300/?tokyo"},

]


for place in places:
    popup_content = f"<b>{place['name']}</b><br><img src='{place['image_url']}' width='200px'>"
    folium.Marker(location=place["coords"], popup=folium.Popup(html=popup_content, max_width='auto')).add_to(world_map)

display(world_map)
```
