```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age_groups = ['18-25', '26-35', '36-45', '46-55', '56-65', '66+']


total_loans = np.random.randint(50, 200, size=len(age_groups))
bad_loans = np.random.randint(0, total_loans)
good_loans = total_loans - bad_loans
percent_bad_loans = (bad_loans / total_loans) * 100
percent_good_loans = 100 - percent_bad_loans
```

```python
data = {
    'Age Group': age_groups,
    'Total Loans': total_loans,
    'Bad Loans': bad_loans,
    'Good Loans': good_loans,
    'Percentage Bad Loans': percent_bad_loans,
    'Percentage Good Loans': percent_good_loans
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(2, 2, 1)
plt.bar(df['Age Group'], df['Total Loans'], color='skyblue')
plt.title('Total Loans')
plt.xlabel('Age Group')
plt.ylabel('Number of Loans')
```

```python
plt.subplot(2, 2, 2)
plt.pie(df['Percentage Bad Loans'], labels=df['Age Group'], autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'orange', 'yellowgreen', 'lightskyblue', 'violet', 'lightpink'])
plt.axis('equal')
plt.title('Percentage of Bad Loans')
```

```python
plt.subplot(2, 2, 3)
plt.pie(df['Percentage Good Loans'], labels=df['Age Group'], autopct='%1.1f%%', startangle=140, colors=['lightskyblue', 'violet', 'lightpink', 'lightcoral', 'orange', 'yellowgreen'])
plt.axis('equal')
plt.title('Percentage of Good Loans')
```

```python
plt.subplot(2, 2, 4)
plt.hist(df['Total Loans'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Histogram of Total Loans')
plt.xlabel('Number of Loans')
plt.ylabel('Frequency')
```
