counting and plotting the frequency of stopwords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from collections import Counter
import matplotlib.pyplot as plt

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)
    return new_string

input_string = "This is an example sentence to remove stop words from."
result = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result)

stop_words = set(stopwords.words('english'))
words = input_string.split()
stopword_freq = Counter([word for word in words if word.lower() in stop_words])

plt.bar(stopword_freq.keys(), stopword_freq.values())
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.title('Frequency of Stopwords')
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
nltk.download('stopwords')

def remove_stop_words(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    new_string = ' '.join(filtered_words)  # Add space between words
    return new_string

def plot_stopword_frequency(string):
    stop_words = set(stopwords.words('english'))
    words = string.split()
    stopwords_count = sum(1 for word in words if word.lower() in stop_words)
    non_stopwords_count = len(words) - stopwords_count

    # Plotting
    labels = ['Stopwords', 'Non-Stopwords']
    values = [stopwords_count, non_stopwords_count]
    plt.bar(labels, values, color=['blue', 'green'])
    plt.xlabel('Word Type')
    plt.ylabel('Frequency')
    plt.title('Frequency of Stopwords vs Non-Stopwords')
    plt.show()

input_string = "This is an example sentence to remove stop words from."
result = remove_stop_words(input_string)
print("Original string:", input_string)
print("Modified string:", result)

plot_stopword_frequency(input_string)
```
